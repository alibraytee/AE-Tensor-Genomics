```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import cluster
from lifelines import KaplanMeierFitter

def k_mean(old_data,tucker_old_data,k = 3,pca = True):
    data = old_data.copy(deep=True)
    newData = tucker_old_data.copy()
    if pca:
        ss = StandardScaler()
        mm = MinMaxScaler()
        tucker_data = ss.fit_transform(tucker_old_data)
        pca=PCA(n_components = 'mle')
        newData=pca.fit_transform(tucker_data)

    k_model = cluster.KMeans(n_clusters=k, max_iter=100, n_jobs=4, init="k-means++")
    k_model.fit(newData)
    
    group = k_model.labels_.T
    
    data.insert(data.shape[1], 'group', group)
    
    plt.style.use('seaborn')
    
    #SA
    
    data.dropna(subset=['overall_survival'], inplace=True)
#     print(data[data['overall_survival'].isnull()])
    data['status'] = [1 if x == 1 else 0 for x in data['status']]
    # df['Overall_survival'] = [x for x in df['overall_survival']]

    T = data['overall_survival']
    E = data['status']
     
    ax = plt.subplot(211)
    kmf = KaplanMeierFitter()
    
    tt=[]
    ee=[]
    
    
    plt.title('Kaplan-Meier Curve')
    for name, grouped_df in data.groupby('group'):
        tt.append(grouped_df['overall_survival'])
        ee.append(grouped_df['status'])
        kmf.fit(grouped_df['overall_survival'], grouped_df['status'], label=name)
        kmf.plot_survival_function(ax=ax)
        
    data_2000 = data.drop(data[data['overall_survival']>2000].index)
    
    ax = plt.subplot(212)
    kmf = KaplanMeierFitter()
    plt.title('Kaplan-Meier Curve less 2000 days')
    for name, grouped_df in data_2000.groupby('group'):
        tt.append(grouped_df['overall_survival'])
        ee.append(grouped_df['status'])
        kmf.fit(grouped_df['overall_survival'], grouped_df['status'], label=name)
        kmf.plot_survival_function(ax=ax)

    #logrank_test 
    from lifelines.statistics import logrank_test 
    results=logrank_test(tt[0],tt[1],event_observed_A= ee[0],event_observed_B=ee[1]) 
    results.print_summary()
    
    results=logrank_test(tt[2],tt[3],event_observed_A= ee[2],event_observed_B=ee[3]) 
    results.print_summary()
          
    return data,data_2000
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import cluster
from lifelines import KaplanMeierFitter
from sklearn import mixture

def OGMM(old_data,tucker_old_data):
    data = old_data.copy(deep=True)
    ss = StandardScaler()
    mm = MinMaxScaler()
    tucker_data = ss.fit_transform(tucker_old_data)
    pca=PCA(n_components = 'mle')
    newData=pca.fit_transform(tucker_data)
    
#     k_model = cluster.KMeans(n_clusters=3, max_iter=100, n_jobs=4, init="k-means++")
#     k_model.fit(newData)
    gmm = mixture.GaussianMixture(n_components=3, covariance_type='full')
    
#     print(gmm.weights_)
    group = gmm.fit_predict(newData).T
    
    data.insert(data.shape[1], 'group', group)
    
    plt.style.use('seaborn')
    
    #SA
    
    data.dropna(subset=['overall_survival'], inplace=True)
#     print(data[data['overall_survival'].isnull()])
    data['status'] = [1 if x == 1 else 0 for x in data['status']]
    # df['Overall_survival'] = [x for x in df['overall_survival']]

    T = data['overall_survival']
    E = data['status']
     
    ax = plt.subplot(211)
    kmf = KaplanMeierFitter()
    plt.title('Kaplan-Meier Curve')
    for name, grouped_df in data.groupby('group'):
        kmf.fit(grouped_df['overall_survival'], grouped_df['status'], label=name)
        kmf.plot_survival_function(ax=ax)
        
    data_2000 = data.drop(data[data['overall_survival']>2000].index)
    
    ax = plt.subplot(212)
    kmf = KaplanMeierFitter()
    plt.title('Kaplan-Meier Curve less 2000 days')
    for name, grouped_df in data_2000.groupby('group'):
        kmf.fit(grouped_df['overall_survival'], grouped_df['status'], label=name)
        kmf.plot_survival_function(ax=ax)
        
    return data,data_2000
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import cluster
from lifelines import KaplanMeierFitter
from sklearn import mixture
from sklearn.cluster import AgglomerativeClustering

def AGG(old_data,tucker_old_data):
    data = old_data.copy(deep=True)
    ss = StandardScaler()
    mm = MinMaxScaler()
    tucker_data = ss.fit_transform(tucker_old_data)
    pca=PCA(n_components = 'mle')
    newData=pca.fit_transform(tucker_data)
    
#     k_model = cluster.KMeans(n_clusters=3, max_iter=100, n_jobs=4, init="k-means++")
#     k_model.fit(newData)
#     gmm = mixture.GaussianMixture(n_components=3, covariance_type='full')
    agg = AgglomerativeClustering(n_clusters=3).fit(newData)

#     print(gmm.weights_)
    group = agg.labels_.T
    
    data.insert(data.shape[1], 'group', group)
    
    plt.style.use('seaborn')
    
    #SA
    
    data.dropna(subset=['overall_survival'], inplace=True)
#     print(data[data['overall_survival'].isnull()])
    data['status'] = [1 if x == 1 else 0 for x in data['status']]
    # df['Overall_survival'] = [x for x in df['overall_survival']]

    T = data['overall_survival']
    E = data['status']
     
    ax = plt.subplot(211)
    kmf = KaplanMeierFitter()
    plt.title('Kaplan-Meier Curve')
    for name, grouped_df in data.groupby('group'):
        kmf.fit(grouped_df['overall_survival'], grouped_df['status'], label=name)
        kmf.plot_survival_function(ax=ax)
        
    data_2000 = data.drop(data[data['overall_survival']>2000].index)
    
    ax = plt.subplot(212)
    kmf = KaplanMeierFitter()
    plt.title('Kaplan-Meier Curve less 2000 days')
    for name, grouped_df in data_2000.groupby('group'):
        kmf.fit(grouped_df['overall_survival'], grouped_df['status'], label=name)
        kmf.plot_survival_function(ax=ax)
        
    return data,data_2000
```
