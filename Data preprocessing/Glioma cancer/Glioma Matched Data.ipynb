```python
# Import some external libraries
import pandas as pd
import numpy as np                 
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading data into dataframe
clinical = pd.read_table('/Users/vivit/Capstone/Glioma/clinical.txt')
methylation = pd.read_table('/Users/vivit/Capstone/Glioma/methy.txt')
miRNASeq = pd.read_table('/Users/vivit/Capstone/Glioma/miRNA.txt')
rnaSeq = pd.read_table('/Users/vivit/Capstone/Glioma/rnaSeq.txt')
scnv = pd.read_table('/Users/vivit/Capstone/Glioma/SCNV.txt')
```

```python
#1090 rows × 72 columns
scnv_t = scnv.T
scnv_t = scnv_t[1:]
scnv_t.columns = scnv['attrib_name']
#512 rows × 791 columns
miRNASeq_t = miRNASeq.T
miRNASeq_t = miRNASeq_t[1:]
miRNASeq_t.columns = miRNASeq['attrib_name']
#669 rows × 20118 columns
rnaSeq_t = rnaSeq.T
rnaSeq_t = rnaSeq_t[1:]
rnaSeq_t.columns = rnaSeq['attrib_name']
#656 rows × 336630 columns

methylation_t = methylation.T
methylation_t = methylation_t[1:]
methylation_t.columns = methylation['attrib_name']

#1097 rows × 20 columns
clinical_t = clinical.T
clinical_t = clinical_t[1:]
clinical_t.columns = clinical['attrib_name']
```

```python
result = miRNASeq_t.join(methylation_t, how="inner")# 755, 783
result
```

```python
result2 = result.join(scnv_t, how="inner") # 623, 1080
result2
```

```python
result3 = result2.join(rnaSeq_t, how="inner")
result3
```

```python
result_final = result2.join(clinical_t, how="inner")
result_final
```

```python
data_clinical = clinical_t.loc[result_final.index]
data_miRnaSeq = miRNASeq_t.loc[result_final.index]
data_methylation = methylation_t.loc[result_final.index]
data_rnaSeq = rnaSeq_t.loc[result_final.index]
data_scnv = scnv_t.loc[result_final.index]
```

```python
data_clinical.to_csv('/Users/vivit/Capstone/Glioma/data_clinical_G.csv')
data_miRnaSeq.to_csv('/Users/vivit/Capstone/Glioma/data_miRnaSeq_G.csv')
data_methylation.to_csv('/Users/vivit/Capstone/Glioma/data_methylation_G.csv')
data_rnaSeq.to_csv('/Users/vivit/Capstone/Glioma/data_rnaSeq_G.csv')
data_scnv.to_csv('/Users/vivit/Capstone/Glioma/data_scnv_G.csv')
```
