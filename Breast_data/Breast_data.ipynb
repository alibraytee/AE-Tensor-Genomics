```python
import pandas as pd
import numpy as np
```

```python
#reading data into dataframe
scnv = pd.read_table('Human__TCGA_BRCA__BI__SCNA__SNP_6.0__01_28_2016__BI__Focal__Firehose_GISTIC2.cct')
miRNASeq = pd.read_table('Human__TCGA_BRCA__BDGSC__miRNASeq__HS_miR__01_28_2016__BI__Gene__Firehose_RPKM_log2.cct')
clinical = pd.read_table('Human__TCGA_BRCA__MS__Clinical__Clinical__01_28_2016__BI__Clinical__Firehose.tsi')
rnaSeq = pd.read_table('Human__TCGA_BRCA__UNC__RNAseq__HiSeq_RNA__01_28_2016__BI__Gene__Firehose_RSEM_log2.cct')
methylation = pd.read_table('Human__TCGA_BRCA__JHU_USC__Methylation__Meth450__01_28_2016__BI__CpG__Firehose_Methylation_Prepocessor.cct')
```

```python
# transpose all dataframe and update headers

#1080 rows × 69 columns
scnv_t = scnv.T
scnv_t = scnv_t[1:]
scnv_t.columns = scnv['attrib_name']
#755 rows × 823 columns
miRNASeq_t = miRNASeq.T
miRNASeq_t = miRNASeq_t[1:]
miRNASeq_t.columns = miRNASeq['attrib_name']
#1093 rows × 20155 columns
rnaSeq_t = rnaSeq.T
rnaSeq_t = rnaSeq_t[1:]
rnaSeq_t.columns = rnaSeq['attrib_name']
#783 rows × 335854 columns
methylation_t = methylation.T
methylation_t = methylation_t[1:]
methylation_t.columns = methylation['attrib_name']

#1097 rows × 20 columns
clinical_t = clinical.T
clinical_t = clinical_t[1:]
clinical_t.columns = clinical['attrib_name']

```

```python
# get the indexes of shared samples
result = miRNASeq_t.join(methylation_t, how="inner")# 755
result2 = result.join(scnv_t, how="inner") # 623
result3 = result2.join(rnaSeq_t, how="inner") # 618
result_final = result3.join(clinical_t, how="inner") # 616
result_final
```

```python
# create the selected data for each techs and clinical
data_clinical = clinical_t.loc[result_final.index]
data_miRnaSeq = miRNASeq_t.loc[result_final.index]
data_methylation = methylation_t.loc[result_final.index]
data_rnaSeq = rnaSeq_t.loc[result_final.index]
data_scnv = scnv_t.loc[result_final.index]
```

```python
#saving data into csv
data_clinical.to_csv('data_clinical.csv')
data_miRnaSeq.to_csv('data_miRnaSeq.csv')
data_methylation.to_csv('data_methylation.csv')
data_rnaSeq.to_csv('data_rnaSeq.csv')
data_scnv.to_csv('data_scnv.csv')
```
